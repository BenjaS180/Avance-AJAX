# üßπ Limpieza de Datos con Pandas - Dataset Adult Income

Este notebook est√° dise√±ado para ense√±ar c√≥mo detectar y corregir problemas comunes de calidad de datos usando el dataset **Adult Income** del UCI Machine Learning Repository.

Se abordar√°n:
- Valores faltantes
- Datos inconsistentes
- Codificaci√≥n de variables categ√≥ricas
- Eliminaci√≥n de duplicados


```python
import pandas as pd
import numpy as np

# Cargar el dataset desde UCI
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data"
column_names = [
    'age', 'workclass', 'fnlwgt', 'education', 'education-num',
    'marital-status', 'occupation', 'relationship', 'race', 'sex',
    'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'income'
]
df = pd.read_csv(url, names=column_names, na_values="?", sep=",", skipinitialspace=True)

# Mostrar las primeras filas
df.head()
```

```python
# Ver informaci√≥n general del dataset
df.info()
```

## üìå Revisi√≥n de valores faltantes

Los valores faltantes han sido cargados como `NaN` gracias al par√°metro `na_values="?"`. Vamos a visualizarlos.


```python
# Contar valores faltantes por columna
df.isnull().sum()
```

## üìå Eliminar o imputar valores faltantes

Decidiremos si eliminamos filas con datos faltantes o si imputamos seg√∫n corresponda.


```python
# Eliminar filas con valores faltantes
df_clean = df.dropna()

# Confirmar que no hay valores nulos
df_clean.isnull().sum()
```

## üìå Normalizaci√≥n de categor√≠as

Vamos a revisar las columnas categ√≥ricas y corregir espacios, may√∫sculas, etc.


```python
# Eliminar espacios innecesarios en columnas categ√≥ricas
cat_cols = df_clean.select_dtypes(include='object').columns

for col in cat_cols:
    df_clean[col] = df_clean[col].str.strip()

# Verificar valores √∫nicos de una columna categ√≥rica
df_clean['native-country'].unique()
```

## üìå Eliminar duplicados (si existen)


```python
# Contar duplicados
df_clean.duplicated().sum()

# Eliminar duplicados
df_clean = df_clean.drop_duplicates()
```
